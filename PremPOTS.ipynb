```python
import requests
from bs4 import BeautifulSoup
from io import StringIO
import pandas as pd
import time

standings_url = "https://en.wikipedia.org/wiki/Premier_League_Player_of_the_Season"
data = requests.get(standings_url)
scrapedData = data.text
soup = BeautifulSoup(data.text)
standings_table= soup.select('table.sortable')[0]
pots = pd.read_html(StringIO(scrapedData), match="Premier League Player of the Season winners")[0] #match es para que agarre la tabla que quieres
```

```python
del pots["Position[A]"]
del pots["Ref(s)"]
```

```python
pots["Club"] = pots["Club"].str.replace("§","",regex = False)
```

```python
pots["Player"] = pots["Player"].str.replace(r'\s*\(.*\)|#.*|†.*|‡.*', '', regex=True) #the first part matches everything after a space in the name
                                                                                      #the second one is if there is a special character after the name
```

```python
pots["Season"] = pots["Season"].str.extract(r'(\d{4})').astype(int) + 1 #captures the first four digits of the year, converts it to int and then adds 1
```

```python
pots = pots.rename(columns={'Club': 'Squad'})
pots.head(30)
```

```python
pots["Winner"] = 1 #el 1 es que si gano en binario pero para no poner yes y despues convertir a int lo hago de una vez
pots.head(30)
```

```python
premDf = pots
premDf.to_csv("POTSWinner.csv")
```
